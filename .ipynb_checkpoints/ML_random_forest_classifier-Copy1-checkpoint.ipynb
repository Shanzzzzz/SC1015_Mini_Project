# Random Forest Classifier

### Importing essential libraries

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from pprint import pprint

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
```

### Importing Datasets (Tomek links, SMOTETOMEK, SMOTEENN)

```python
tomek_data = pd.read_csv("Data/tomek_data.csv", index_col=0)
tomek_data.head()
```

```python
tomek_data.describe()
```

```python
smotetomek_data = pd.read_csv("Data/smotetomek_data.csv", index_col = 0)
smotetomek_data.head()
```

```python
smotetomek_data.describe()
```

```python
smoteenn_data = pd.read_csv("Data/smoteenn_data.csv", index_col = 0)
smoteenn_data.head()
```

```python
smoteenn_data.describe()
```

```python
tomek_data      = tomek_data.rename(columns = {'Coronary Heart Disease/Myocardial Infarction':'CHD/MI'})
smotetomek_data = smotetomek_data.rename(columns = {'Coronary Heart Disease/Myocardial Infarction':'CHD/MI'})
smoteenn_data   = smoteenn_data.rename(columns = {'Coronary Heart Disease/Myocardial Infarction':'CHD/MI'})
```

```python
tomek_data      = tomek_data.rename(columns = {'No of Days of Poor Mental Health ':'Mental Health',
                                               'No of Days of Poor Physical Health':'Physical Health',
                                               'Total Sleeping Time':'Total Sleep'})
smotetomek_data = smotetomek_data.rename(columns = {'No of Days of Poor Mental Health ':'Mental Health',
                                               'No of Days of Poor Physical Health':'Physical Health',
                                               'Total Sleeping Time':'Total Sleep'})
smoteenn_data   = smoteenn_data.rename(columns = {'No of Days of Poor Mental Health ':'Mental Health',
                                               'No of Days of Poor Physical Health':'Physical Health',
                                               'Total Sleeping Time':'Total Sleep'})
```

### Import test data

```python
test_data = pd.read_csv('Data/test.csv', index_col=0)
test_data = test_data.rename(columns = {'Coronary Heart Disease/Myocardial Infarction':'CHD/MI'})
# Setting up test data for model evaluation
test_features = test_data.drop('CHD/MI', axis = 1)
test_labels   = test_data['CHD/MI']
```

### Fitting model with tomek_data

```python
tomek_train_features = tomek_data.drop('CHD/MI', axis = 1)
tomek_train_labels   = tomek_data['CHD/MI']
```

```python
rf = RandomForestClassifier(random_state = 516)

rf.fit(tomek_train_features, tomek_train_labels)
```

```python
tomek_train_pred = rf.predict(tomek_train_features)
test_pred        = rf.predict(test_features) 

# Check the Goodness of Fit (Train)
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", rf.score(tomek_train_features, tomek_train_labels))
print()

# Check the Goodness of Fit (Test)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", rf.score(test_features, test_labels))
print()
```

```python
f, axes = plt.subplots(1, 2, figsize=(15, 5))
sb.heatmap(confusion_matrix(tomek_train_labels, tomek_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(test_labels, test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
```

```python
# Print the Accuracy Measures from the Confusion Matrix
cmTrain = confusion_matrix(tomek_train_labels, tomek_train_pred)
tpTrain = cmTrain[1][1] 
fpTrain = cmTrain[0][1] 
tnTrain = cmTrain[0][0] 
fnTrain = cmTrain[1][0] 

print("True Positive Rate Train :\t", round(tpTrain/(tpTrain + fnTrain), 6))
print("True Negative Rate Train :\t", round(tnTrain/(tnTrain + fpTrain), 6))
print()

print("False Positive Rate Train :\t", round(fpTrain/(tnTrain + fpTrain), 6))
print("False Negative Rate Train :\t", round(fnTrain/(tpTrain + fnTrain), 6))
print()

print("=================================================================\n")

cmTest = confusion_matrix(test_labels, test_pred)
tpTest = cmTest[1][1] 
fpTest = cmTest[0][1] 
tnTest = cmTest[0][0] 
fnTest = cmTest[1][0] 

print("True Positive Rate Test :\t", round(tpTest/(tpTest + fnTest), 6))
print("True Negative Rate Test :\t", round(tnTest/(tnTest + fpTest), 6))
print()

print("False Positive Rate Test :\t", round(fpTest/(fpTest + tnTest), 6))
print("False Negative Rate Test :\t", round(fnTest/(fnTest + tpTest), 6))
```

### Fitting model with smoteenn_data

```python
smoteenn_train_features = smoteenn_data.drop('CHD/MI', axis = 1)
smoteenn_train_labels   = smoteenn_data['CHD/MI']
```

```python
rf = RandomForestClassifier(random_state = 516)

rf.fit(smoteenn_train_features, smoteenn_train_labels)
```

```python
smoteenn_train_pred = rf.predict(smoteenn_train_features)
test_pred        = rf.predict(test_features) 

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", rf.score(smoteenn_train_features, smoteenn_train_labels))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", rf.score(test_features, test_labels))
print()
```

```python
f, axes = plt.subplots(1, 2, figsize=(15, 5))
sb.heatmap(confusion_matrix(smoteenn_train_labels, smoteenn_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(test_labels, test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
```

```python
# Print the Accuracy Measures from the Confusion Matrix
cmTrain = confusion_matrix(smoteenn_train_labels, smoteenn_train_pred)
tpTrain = cmTrain[1][1] 
fpTrain = cmTrain[0][1] 
tnTrain = cmTrain[0][0] 
fnTrain = cmTrain[1][0] 

print("True Positive Rate Train :\t", round(tpTrain/(tpTrain + fnTrain), 6))
print("True Negative Rate Train :\t", round(tnTrain/(tnTrain + fpTrain), 6))
print()

print("False Positive Rate Train :\t", round(fpTrain/(tnTrain + fpTrain), 6))
print("False Negative Rate Train :\t", round(fnTrain/(tpTrain + fnTrain), 6))
print()

print("=================================================================\n")

cmTest = confusion_matrix(test_labels, test_pred)
tpTest = cmTest[1][1] # True Positives : Y (1) predicted Y (1)
fpTest = cmTest[0][1] # False Positives : N (0) predicted Y (1)
tnTest = cmTest[0][0] # True Negatives : N (0) predicted N (0)
fnTest = cmTest[1][0] # False Negatives : Y (1) predicted N (0)

print("True Positive Rate Test :\t", round(tpTest/(tpTest + fnTest), 6))
print("True Negative Rate Test :\t", round(tnTest/(tnTest + fpTest), 6))
print()

print("False Positive Rate Test :\t", round(fpTest/(fpTest + tnTest), 6))
print("False Negative Rate Test :\t", round(fnTest/(fnTest + tpTest), 6))
```
